### **Calculating Top 10 highest average rating movies, according to the minimum amount of ratings inputted by the user**

For this section, we took to a more brute force approach of solving the program. Instead of using some of the built-in functions that would have made the process much more simple. The function takes in two parameters which are an empty array and a integer value which indicates the minimum amount of ratings that should be considered. What we did first was create a local variable within the function that held the dataframe of a groupby of the title and genre and then took the mean of those groupings so that we didn't have redundant movies or genre. After that we created another local variable within the function rating_count which held the count of the ratings so that we knew per group how many ratings that movie had. The use of these two variables that held dataframes was so that we didn't have to do another merge and simultaneously when we iterated over one--as we did with our first loop which removes the movies of the average rating according to the passed minimum. What the for loop does is iterate over the ratings_count dataframe and wherever it finds some rating count less than the one passed it will delete it in the average_rating data frame. No work needs to be done on the rating_count function as it will not output any of its contents. The following nested while loop simply adds the top 10 rated movies according to the average ratings of each of the movies. This was done in a brute forced fashion whereby it simply finds the max of the dataframe and then drops that max, so that the following iteration finds the next max. This algorithm also outputs the contents of the list in sorted fashion from greatest to least. Truly like a top 10 list!!!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
ratings_df = ratings_df.drop(columns = 'timestamp')
```

```python
cleaned_data = movies_df.merge(ratings_df, on = 'movieId', how = 'inner') #merging on inner so that it drops null
print(cleaned_data)
print(cleaned_data.isna().sum())#checking for empty.
print(cleaned_data.duplicated().sum())#checking for duplicates. 
#No duplicates
#No empty
```

```python
# Simple statistics - BRANDON
total_movies = movies_df['title'].nunique()
print("The total amount of movies are,", total_movies)#total amount of movies. 
#average number of ratings per movie
avg_num_of_rating_per_movie = cleaned_data.groupby('title')['rating'].count()
print() #whitespace
print("The average number of ratings per movie is,", avg_num_of_rating_per_movie.mean())
```

```python
##Finding all the unique Genre.

genre_column_df = pd.DataFrame(movies_df['genres'])##make a dataframe with just the genres column.

genre_column_df = genre_column_df['genres'].str.split(r'\|') #make little arrays for the multi genre by splitting each word by |
new_genre_column_df = pd.DataFrame(genre_column_df.explode('genres')) #break up each list, and put their contens as individual items of the genre column
new_genre_column_df = new_genre_column_df.drop_duplicates()

for i in new_genre_column_df.index: #drop the "(no genres listed)"
    if (new_genre_column_df.loc[i, 'genres'] == "(no genres listed)"):
        new_genre_column_df = new_genre_column_df.drop(i)
            
new_genre_column_df = new_genre_column_df.sort_values(by = ['genres']) #put the genres in alphabetical order
unique_genre_df = new_genre_column_df['genres']
unique_genre_df.index = np.arange(1, len(unique_genre_df) + 1) #overwrite the previous index as they don't matter. 

print(unique_genre_df)
```

```python
# Average rating per movie. -BRANDON
avg_rating_per_movie = cleaned_data.groupby('title', as_index = False)['rating'].mean()#Keep the dataframe using as_index = false
print(avg_rating_per_movie)
```

```python
##Top 10 highest rated (average rating), according to inputted min. -BRANDON

def top_ten_highest_rated(array_of_max, minimum):
    avg_rating = cleaned_data.groupby('title', as_index = False)['rating'].mean()
    #using as_index = False to get the goupby into a df and not a series.
    rating_count = cleaned_data.groupby('title', as_index = False)['rating'].size()
    #contains the amount of ratings per movie. 
    
    for i in rating_count.index: #drop the ratings grouped by movie that have less than minimum
        if (rating_count.loc[i, 'size'] < minimum):
            avg_rating = avg_rating.drop(i)
    
    while(len(array_of_max) < 10 and (not avg_rating.empty)):
        #Find and add the 10 top highest ratings sorted from highest ratings to lowest.
        #not .empty is in case there aren't 10 items to add to the array, this could occur with a very large min.
        for i in avg_rating.index:
            if len(array_of_max) == 10:
                #Since we iterate over the entire df, we need to make sure we break when we reach len 10 for the array
                #Sometimes max could occurr twice in the same call of the for loop.
                break
            elif (avg_rating.loc[i, 'rating'] == avg_rating['rating'].max()):
                array_of_max.append(avg_rating.loc[i, 'title'])#add item by title to array.
                avg_rating = avg_rating.drop(i)#drop it so we don't add the same max.
    return array_of_max
```

```python
#Identify the top 10 most-rated movies (popularity, number of ratings ). -BALJINDER

df = cleaned_data.copy()
# Count the number of instances of each movieId and creates count column
title_counts = df.groupby('title').size().reset_index(name='count')

# Merge the count of instances into the movies DataFrame
movies = pd.read_csv('movies.csv')
# Merge the count of instances into the movies DataFrame
movies_with_counts = pd.merge(movies, title_counts, on='title')

# Output the titles of the 10 most-rated movies
most_rated = movies_with_counts.nlargest(10, 'count')
most_rated[['title', 'count']]

##Making a bar plot, to visualize the top 10, using the .plot.bar(...)
most_rated.plot.bar(x = 'title', y = 'count', title='Top 10 most rated movies', figsize=(10,5), fontsize=12, align = 'center')

plt.show()
```

```python
#Analyze the distribution of ratings across different genres. (it is an open question)-BALJINDER

data = cleaned_data.copy() #copy the merged data frame
data['genres'] = data['genres'].str.split(r'\|') #split all the multi-genre into arrays by the '|' character 
exploded_df = data.explode('genres') #Explode those individual arrays so that they are now 
exploded_df = exploded_df[exploded_df['genres'] != '(no genres listed)']
average_ratings_per_genre = exploded_df.groupby('genres')['rating'].mean()


print(average_ratings_per_genre)
```

```python
#Printing the highest average rating genre and lowest. 
print("The highest average rating out of the genre is,", average_ratings_per_genre.idxmax(), "with an average rating of, ", average_ratings_per_genre.max())
print("The lowest average rating out of the genre is,", average_ratings_per_genre.idxmin(), "with an average rating of, ", average_ratings_per_genre.min())
```

```python
# copy the dataset 
cleaned_copy_1 = cleaned_data.copy()
# Extract the release_year from the title column using a regular expression
cleaned_copy_1['year'] = cleaned_copy_1['title'].str.extract(r'\((\d{4})\)')
cleaned_copy_1 = cleaned_copy_1[cleaned_copy_1['genres'] != '(no genres listed)'] #remove "(no genres listed)"
cleaned_copy_1 = cleaned_copy_1.dropna() 
#before we convert we need to drop Na, since there may be some NULL titles that were extracted since,
#the title might have not had a date
cleaned_copy_1['genres'] = cleaned_copy_1['genres'].str.split(r'\|') #make little arrays for the multi genre by splitting each word by |
cleaned_copy_1 = cleaned_copy_1.explode('genres')
# Convert the release_year column to integer data type
cleaned_copy_1['year'] = cleaned_copy_1['year'].astype(int)
# Group the data by genre and release year
cleaned_grouped_1 = cleaned_copy_1.groupby(['genres', 'year'])
#mean_ratings = df.groupby(['genre', 'release_year'])['rating'].mean() can merge the line. 
cleaned_mean_ratings = cleaned_grouped_1.mean()['rating']#get the mean for 
# Reset the index to convert the multi-indexed dataframe to a flat dataframe
cleaned_mean_ratings = cleaned_mean_ratings.reset_index()

print(cleaned_mean_ratings)
```
